```python
#quiz game with timer
import random
from sys import exit
import time
import numpy as n

def answerScrambler(random_question):
    if len(random_question) != 7:
        print('the question is not formatted correctly, turning off...')
        exit()
    correct_index = (random_question[6] - 1)
    random_question.pop(6)
    question = random_question[0]
    random_question.pop(0)
    correct_answer = random_question[correct_index]
    random_question.pop(correct_index)
    
    i = 0
    tries = 3
    scrambled_answers = []
    for i in range(4):
        random_number = random.randint(0, tries)
        wrong_answer = random_question[random_number]
        scrambled_answers.append(wrong_answer)
        random_question.pop(random_number)
        tries = tries - 1
    
    random_index = random.randint(0, 5)
    correct_index = random_index
    scrambled_answers.insert(random_index, correct_answer)
    exit_list = [question, scrambled_answers, correct_index]
    return exit_list
```

```python
def abc_er(question_bundle, abc):
    answers = question_bundle[1]
    for i in range(len(answers)):
        answer = answers[i]
        letter = abc[i]
        final = letter + answer
        answers.pop(i)
        answers.insert(i, final)
    return answers
```

```python
def questionAsker(abc_ed_answers, question_bundle):
    question = question_bundle[0]
    i = 0
    print(question)
    for i in range(len(abc_ed_answers)):
        answer = abc_ed_answers[i]
        print(answer)
```

```python
def isCorrect(person_input, question_bundle):
    taketry = 0
    add_point = 0
    abc2 = ['A', 'B', 'C', 'D', 'E']
    correct_index = question_bundle[2]
    answers = question_bundle[1]
    try:
        person_index = abc2.index(person_input)
    except:
        person_index = 10   
    if person_index < 10:
        if person_index == correct_index:
            print('that is correct!!!')
            print('')
            add_point = add_point + 1
            taketry = taketry + 1
        else:
            print('better luck next time.')
            print('')
    else:
        print('that was not an answer, question skipped.')
        print('')
    wholelist = [add_point, taketry]
    return wholelist
```

```python
q1 = ['what is 1 + 1?', '0', '1', '2', '3', '4', 3]
q2 = ['how much does Ethan weigh in kilos?', '36','20','150','2','What is a kilo?', 1]
q3 = ['in quantum mechanics is a photon a particle or a wave?', 'particle', 'wave', 'neither', 'both', 'what is quantum mechanics?', 4]
q4 = ['when is a door not a door?', 'whan it is a jar', 'when it is a teapot', 'when it is an elephant', 'when it is a nutcracker', 'when it is a pudding', 1]
q5 = ['what comes after \'q\'?', 'w', 'd', 'y', 'l', 'f', 1, 'hard']
q6 = ['what is the meaning of life?', '42', 'i dunno', '12', 'to live with purpose', 'to live with porpoises', 1]
questions = [q1, q2, q3, q4, q5, q6]
```

```python
import random
shuffledQuestions = []
i = 0
tries = ((len(questions)) - 1)
for i in range(len(questions)):
    random_number = random.randint(0, tries)
    question = questions[random_number]
    shuffledQuestions.append(question)
    questions.pop(random_number)
    tries = tries - 1
```

```python
print('you get 1 point for each of the ', len(shuffledQuestions), ' that you get right. Good luck.')
print('')
times = len(shuffledQuestions)
points = 0
i = 0
for i in range(len(shuffledQuestions)):
    abc = ['A - ', 'B - ', 'C - ', 'D - ', 'E - ']
    print('question ', i + 1, ':')
    print('')
    random_question = shuffledQuestions[i]
    question_bundle = answerScrambler(random_question)
    abc_ed_answers = abc_er(question_bundle, abc)
    questionAsker(abc_ed_answers, question_bundle)
    person_input = input('what is your answer?, (A, B, C, D, E)')
    print('')
    addpoint_lesstries = isCorrect(person_input, question_bundle)
    add_point = addpoint_lesstries[0]
    lesstries = addpoint_lesstries[1]
    times = times - lesstries
    if times == 0:
        i = (len(shuffledQuestions) - 1)
    
    if i == (len(shuffledQuestions) - 1) and times > 0:
        i = 0
    points = points + add_point
    time.sleep(1)
print('you got ', points, ' out of ', len(shuffledQuestions), ' possible points!')
if points < (len(shuffledQuestions) // 2):
    print('you suck')
else:
    print('well done')
print('')
print('end')
```
