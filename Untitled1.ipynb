```python
from panda3d.core import Point3
from direct.showbase.ShowBase import ShowBase

class MyApp(ShowBase):
    def __init__(self):
        super().__init__()  # Initialize the base class

        # Load a 3D model (you can replace this with your own model)
        self.model = self.loader.loadModel("models/panda")  # 'panda' is a built-in model
        self.model.reparentTo(self.render)  # Re-parent it to the render tree (scene)

        # Scale and position the model
        self.model.setScale(0.5, 0.5, 0.5)
        self.model.setPos(0, 10, 0)

        # Add a light source to the scene
        self.addLight()

    def addLight(self):
        from panda3d.core import DirectionalLight
        from panda3d.core import Point3

        # Create a light and attach it to the scene
        light = DirectionalLight('light')
        lightNode = self.render.attachNewNode(light)

        # Set the light position and direction
        lightNode.setPos(Point3(10, 10, 10))
        lightNode.lookAt(Point3(0, 0, 0))

        # Attach the light to the scene graph
        self.render.setLight(lightNode)


# Start the app
app = MyApp()
app.run()
```
