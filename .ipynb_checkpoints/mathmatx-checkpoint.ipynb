```python
import time as t
import os
import random

class primes:
    class primelist:
        def isPrimeNumber(n1):
            n2 = 1
            count = 0
            while n2 <= n1:
                if n1 % n2 == 0:
                    count = count + 1
                n2 = n2 + 1
                if count > 2:
                    return False
            if count == 2:
                return True
        
        def primelist(start, end):
            n1 = start
            target = end
            prime_no_list = []
            i = 0
            while n1 <= target:#this ends the program when a spicific number is reached
            #while n1 <= n1:#this keeps going for ever
                trueORfalse = primes.primelist.isPrimeNumber(n1)
                if trueORfalse:
                    prime_no_list.append(n1)
                    #print (n1)
                    #t.sleep(0.1)
                n1 = n1 + 1
            return prime_no_list

    def factors(number):
        smallerNumber = number
        n1 = 2
        listONos = []
        while smallerNumber > 1:
            if smallerNumber % n1 == 0:
                listONos.append(n1)
                smallerNumber = smallerNumber // n1
            else:
                n1 += 1
        return listONos
```
