```python
def encoding(n, e, c):
    return (pow(c, e) % n)
```

```python
def decoding(a, p, q, e, d=None):
    n = p*q
    if d == None:
        lookingForD = True
        counter = 1
        while lookingForD:
            d = counter
            if ((e*d) % ((p-1)*(q-1))) == 1:
                lookingForD = False
                print(f'd is {d}')
            else:
                counter += 2
        return pow(a, d) % n
```
