```python
import random
from sys import exit

amount1 = input('how many dice do you want to roll as an integer')
try:
    amount = int(amount1)
except:
    print('PLEASE WRITE IT AS AN INTEGER')
    exit()

if amount > 10000:
    print('please lower your number')
    exit()

def randomFinder():
    randomNumber = random.randint(0, 6)
    return randomNumber

enter = input('press enter when you want to start rolling the dice')

if enter != '':
    print('please only type enter')
    exit()

enter2 = input('press enter when you want to stop rolling the dice')

if enter2 != '':
    print('please only type enter')
    exit()

for i in range(amount):
    randomNumber = randomFinder()
    print('your number', (i + 1), ' random dice roll is...   ', randomNumber)
```
